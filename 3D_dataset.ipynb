```python
#加入归一化

import os
import numpy as np
import nibabel as nib
import matplotlib.pyplot as plt

def save_slices_from_nii(nii_path, output_folder):
    nii_img = nib.load(nii_path)
    data = nii_img.get_fdata()

    # 获取NIfTI文件的基本名字（不带路径和后缀）
    base_name = os.path.basename(nii_path).replace(".nii.gz", "")
    
    # 为当前NIfTI文件创建子文件夹
    current_output_folder = os.path.join(output_folder, base_name)
    if not os.path.exists(current_output_folder):
        os.makedirs(current_output_folder)

    # 归一化函数
    def normalize_image(img):
        img_min = np.min(img)
        img_max = np.max(img)
        return ((img - img_min) / (img_max - img_min) * 255).astype(np.uint8)

    # 保存切片到子文件夹
    for i in range(data.shape[0]):
        slice_data = normalize_image(data[i, :, :])
        output_path = os.path.join(current_output_folder, f'slice_{i:03}.png')
        plt.imsave(output_path, slice_data, cmap='gray')

# 文件夹路径
input_folder = "/home/xulei/projects/wei/2DSAM3D/dataset/3d_images"
output_folder = "/home/xulei/projects/wei/2DSAM3D/dataset/3_2_norm"

# 如果输出文件夹不存在，则创建
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# 遍历文件夹并处理每个.nii.gz文件
for filename in os.listdir(input_folder):
    if filename.endswith(".nii.gz"):
        nii_path = os.path.join(input_folder, filename)
        save_slices_from_nii(nii_path, output_folder)
```

```python
#加入归一化 保存为两个文件夹


import os
import numpy as np
import nibabel as nib
import matplotlib.pyplot as plt

def save_slices_from_nii(nii_path, output_base_folder):
    nii_img = nib.load(nii_path)
    data = nii_img.get_fdata()

    # 获取NIfTI文件的基本名字（不带路径和后缀）
    base_name = os.path.basename(nii_path).replace(".nii.gz", "")
    
    # 根据文件名确定是图像还是掩膜，并选择相应的输出文件夹
    if "IMG" in base_name:
        output_folder = os.path.join(output_base_folder, "IMG")
    elif "MASK" in base_name:
        output_folder = os.path.join(output_base_folder, "MASK")
    else:
        raise ValueError("Unknown file type. Expecting 'image' or 'mask' in the filename.")

    # 归一化函数
    def normalize_image(img):
        img_min = np.min(img)
        img_max = np.max(img)
        return ((img - img_min) / (img_max - img_min) * 255).astype(np.uint8)

    # 保存切片到文件夹
    for i in range(data.shape[0]):
        slice_data = normalize_image(data[i, :, :])
        output_path = os.path.join(output_folder, f'{base_name}_{i:03}.png')
        plt.imsave(output_path, slice_data, cmap='gray')

# 文件夹路径
input_folder = "/home/xulei/projects/wei/2DSAM3D/dataset/3d_images"
output_base_folder = "/home/xulei/projects/wei/2DSAM3D/dataset/3_2_norm_2"

# 创建图像和掩膜的输出文件夹
image_output_folder = os.path.join(output_base_folder, "IMG")
mask_output_folder = os.path.join(output_base_folder, "MASK")
if not os.path.exists(image_output_folder):
    os.makedirs(image_output_folder)
if not os.path.exists(mask_output_folder):
    os.makedirs(mask_output_folder)

# 遍历文件夹并处理每个.nii.gz文件
for filename in os.listdir(input_folder):
    if filename.endswith(".nii.gz"):
        nii_path = os.path.join(input_folder, filename)
        save_slices_from_nii(nii_path, output_base_folder)




```

```python
import os
import numpy as np
import nibabel as nib
from PIL import Image

def save_slices_to_nii(image_folder, output_folder):
    slices = []

    # 按文件名排序确保图像是按正确的顺序加载的
    filenames = sorted(os.listdir(image_folder))

    # 遍历子文件夹中的所有.png图片
    for filename in filenames:
        if filename.endswith('.png'):
            img_path = os.path.join(image_folder, filename)
            img = Image.open(img_path)
            img_array = np.array(img)
            slices.append(img_array)

    # 将所有的图片堆叠起来形成一个3D数组
    combined_data = np.stack(slices, axis=0)

    # 使用 nibabel 创建 NIfTI 图像并保存
    nii_img = nib.Nifti1Image(combined_data, np.eye(4))
    output_path = os.path.join(output_folder, f'{os.path.basename(image_folder)}.nii.gz')
    nib.save(nii_img, output_path)

# 文件夹路径
input_folder = "/home/xulei/projects/wei/2DSAM3D/dataset/infer_div"
output_folder = "/home/xulei/projects/wei/2DSAM3D/dataset/combined_niis"

# 如果输出文件夹不存在，则创建
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# 遍历文件夹并处理每个子文件夹
for foldername in os.listdir(input_folder):
    current_folder_path = os.path.join(input_folder, foldername)
    if os.path.isdir(current_folder_path):
        save_slices_to_nii(current_folder_path, output_folder)
```

```python
# 把分出来的mask组合回去

import os
import numpy as np
import nibabel as nib
from PIL import Image

def save_slices_to_nii(image_folder, output_folder, original_nii_path):
    slices = []
    
    # 从原始文件中获取头部和仿射矩阵信息
    original_nii = nib.load(original_nii_path)
    affine = original_nii.affine
    header = original_nii.header

    # 按文件名排序确保图像是按正确的顺序加载的
    filenames = sorted(os.listdir(image_folder))

    # 遍历子文件夹中的所有.png图片
    for filename in filenames:
        if filename.endswith('.png'):
            img_path = os.path.join(image_folder, filename)
            img = Image.open(img_path)
            
            # Resize the image to 512x512
#             img = img.resize((512, 512), Image.LANCZOS)

            img_array = np.array(img)
            slices.append(img_array)

    # 将所有的图片堆叠起来形成一个3D数组
    combined_data = np.stack(slices, axis=0)

    # 使用 nibabel 创建 NIfTI 图像并保存
    nii_img = nib.Nifti1Image(combined_data, affine, header=header)
    output_path = os.path.join(output_folder, f'{os.path.basename(image_folder)}.nii.gz')
    nib.save(nii_img, output_path)

# 文件夹路径
input_folder = "/home/xulei/projects/wei/2DSAM3D/dataset/3_2"
output_folder = "/home/xulei/projects/wei/2DSAM3D/dataset/combined_niis_ori_noresize"
original_nii_path = "/home/xulei/projects/wei/2DSAM3D/dataset/3d_images/MASK_0002.nii.gz"

# 如果输出文件夹不存在，则创建
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# 遍历文件夹并处理每个子文件夹
for foldername in os.listdir(input_folder):
    current_folder_path = os.path.join(input_folder, foldername)
    if os.path.isdir(current_folder_path):
        save_slices_to_nii(current_folder_path, output_folder, original_nii_path)
```

```python
# 把分出来的mask组合回去2

import os
import nibabel as nib
import matplotlib.pyplot as plt
import numpy as np

def save_nii_from_slices(input_folder, output_path):
    # 读取所有的PNG文件，并将它们排序
    slice_files = [f for f in os.listdir(input_folder) if f.endswith(".png")]
    slice_files.sort()

    # 读取第一张图片来获取基本信息
    first_img = plt.imread(os.path.join(input_folder, slice_files[0]))
    dtype = first_img.dtype
    # 初始化3D数组
    data = np.zeros((len(slice_files), first_img.shape[0], first_img.shape[1]), dtype=dtype)

    # 读取每张图片，并将其添加到3D数组中
    for idx, slice_file in enumerate(slice_files):
        img_path = os.path.join(input_folder, slice_file)
        # 选择R通道
        data[idx, :, :] = plt.imread(img_path)[:, :, 0]


    # 使用nibabel创建一个新的NIfTI图像，并保存
    nii_img = nib.Nifti1Image(data, affine=np.eye(4))
    nib.save(nii_img, output_path)

# 文件夹路径
input_base_folder = "/home/xulei/projects/wei/2DSAM3D/dataset/3_2"
output_folder = "/home/xulei/projects/wei/2DSAM3D/dataset/combined_niis_ori2"

# 如果输出文件夹不存在，则创建
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# 遍历子文件夹并处理每组PNG文件
for subdir in os.listdir(input_base_folder):
    input_subdir = os.path.join(input_base_folder, subdir)
    if os.path.isdir(input_subdir):
        output_path = os.path.join(output_folder, f"{subdir}.nii.gz")
        save_nii_from_slices(input_subdir, output_path)
```

```python
# 把推理的mask合起来

import os
import nibabel as nib
import matplotlib.pyplot as plt
import numpy as np
from scipy.ndimage import zoom

def resize_image(img, target_shape=(512, 512)):
    """调整图像大小到目标形状。"""
    factors = (target_shape[0] / img.shape[0], target_shape[1] / img.shape[1])
    
    # 如果图像是3D，保持第三维度不变
    if len(img.shape) == 3:
        factors += (1,)
    
    return zoom(img, factors, order=1)

def save_nii_from_slices(input_folder, output_path):
    # 读取所有的PNG文件，并将它们排序
    slice_files = [f for f in os.listdir(input_folder) if f.endswith(".png")]
    slice_files.sort()

    # 读取第一张图片来获取基本信息
    first_img = plt.imread(os.path.join(input_folder, slice_files[0]))
    dtype = first_img.dtype
    # 初始化3D数组，考虑到新的512x512尺寸
    data = np.zeros((len(slice_files), 512, 512), dtype=dtype)

    # 读取每张图片，并将其添加到3D数组中
    for idx, slice_file in enumerate(slice_files):
        img_path = os.path.join(input_folder, slice_file)
        img_data = plt.imread(img_path)
        
        # 将图像左右翻转
        img_data = img_data[:, ::-1]        
        
        # 将图像调整到512x512
        img_data = resize_image(img_data)
        # 如果是3D图像，则取R通道；否则直接赋值
        if len(img_data.shape) == 3:
            data[idx, :, :] = img_data[::-1, :, 0]
        else:
            data[idx, :, :] = img_data[::-1, :]

    # 调整数据的维度，以便它们按照(depth, height, width)的顺序排列
#     data = data.transpose(2, 0, 1)

    # 使用nibabel创建一个新的NIfTI图像，并保存
    nii_img = nib.Nifti1Image(data, affine=np.eye(4))
    nib.save(nii_img, output_path)

# 文件夹路径
input_base_folder = "/home/xulei/projects/wei/2DSAM3D/dataset/infer_div"
output_folder = "/home/xulei/projects/wei/2DSAM3D/dataset/combined_infer_div"

# 如果输出文件夹不存在，则创建
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# 遍历子文件夹并处理每组PNG文件
for subdir in os.listdir(input_base_folder):
    input_subdir = os.path.join(input_base_folder, subdir)
    if os.path.isdir(input_subdir):
        output_path = os.path.join(output_folder, f"{subdir}.nii.gz")
        save_nii_from_slices(input_subdir, output_path)

```

```python
# 合并mask和原图

import os
from PIL import Image

def combine_images(folderA, folderB, folderC):
    # 检查并创建输出文件夹
    if not os.path.exists(folderC):
        os.makedirs(folderC)

    # 对每一对图片进行处理
    for i in range(325):  # 假设你有从0到324，共325张图片
        # 根据提供的命名规则构建文件名
        img_name_A = f"slice_{i:03}.png"
        img_name_B = f"IMG_0002_{i}.png"

        img_path_A = os.path.join(folderA, img_name_A)
        img_path_B = os.path.join(folderB, img_name_B)

        # 读取原图和mask
        img = Image.open(img_path_A)
        mask = Image.open(img_path_B)

        # 将mask resize到512x512
        mask = mask.resize((512, 512))

        # 创建一个RGBA图像，用于放置带透明度的mask
        mask_rgba = Image.new("RGBA", mask.size)
        
        for x in range(mask.width):
            for y in range(mask.height):
                pixel = mask.getpixel((x, y))
                if isinstance(pixel, int):  # 如果像素是整数（灰度图像）
                    mask_rgba.putpixel((x, y), (pixel, pixel, pixel, 128))  # 将灰度值复制到RGB，并添加透明度
                else:
                    mask_rgba.putpixel((x, y), (*pixel, 128))  # 如果是RGB图像，则直接添加透明度

        # 在原图上贴上mask
        img.paste(mask_rgba, (0, 0), mask_rgba)

        # 保存到输出文件夹
        img.save(os.path.join(folderC, img_name_A))

# 使用函数
combine_images('/home/xulei/projects/wei/2DSAM3D/dataset/3_2_norm/IMG_0002', 
               '/home/xulei/projects/wei/2DSAM3D/dataset/infer_div/IMG_0002', 
               '/home/xulei/projects/wei/2DSAM3D/dataset/infer_div/0002')
```
